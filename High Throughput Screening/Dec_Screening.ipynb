```python
import pandas as pd

# 读取表格数据
df = pd.read_csv("pvk_data.csv")
df
```

```python
# 根据条件筛选数据
filtered_df = df[(df['Dec'] < 0.3) & (df['Tolerance factor'] > 0.813) & (df['Tolerance factor'] < 1.107)
                 & (df['Octahedral factor'] > 0.442) & (df['Octahedral factor'] < 0.895)]
filtered_df
```

```python
# 保存筛选后的数据为新的表格文件
filtered_df.to_csv("Dec_Screening.csv", index=False)
```
